```python
def filter_high_scores(list_of_tuples):
    results = {}
    for name, score in list_of_tuples: #unpack each tuple into the variables name and score
        if score > 80 and name.startswith("J"):
            results[name] = score #creates a new entry
            
    return results

#lets test is out
list_of_scores =[("Edwin",10), ("James",84), ("Louise", 90)]

print(filter_high_scores(list_of_scores))

#or we can use a more pythonic dict comprehension approach
def filter_high_scores_py(list_of_tuples):
    return {name: score for name, score in list_of_tuples if score>80 and name.lower().startswith("j")}

print(filter_high_scores_py(list_of_scores))
```

```python
def filter_scores(list_of_list_tuples):
    results={}
    for i, clist in enumerate(list_of_list_tuples):
        print(f" class list {i} is {clist}")
        #now I've got each list of students in each class. lets loop through and add the results to a dict
        for name, score in clist:
            if score > 85:
                results[name] = score
        
    return results    

#now I'll create a dict comprehension version
def filter_score_py(list_of_list_tuples):
    return{name:score for i, clist in enumerate(list_of_list_tuples) for name, score in clist if score > 85}


classrooms = [
    [("Edwin", 78), ("James", 91)],
    [("Sarah", 88), ("John", 95), ("Louise", 60)],
    [("Jake", 83), ("Jo", 79)]
]

print(filter_scores(classrooms))

print(filter_score_py(classrooms))
```

```python
#Classes
class CounterClicker:
    """A class for a counter"""
    
    #this is a dunder (double underscore) method â€” part of Python's special method system, not private. 
    #Python doesn't enforce access restrictions, but a single underscore is the convention for 'internal use'.
    
    def __init__(self, count=0):
        self.count = count
        
    def __repr__(self): #produces the string representation of a class instance
        return f"CounterClicker(count={self.count})"
    
    #implement the public methods
    def click(self, num_times=1):
        self.count += num_times
        
    def read(self):
        return self.count
    
    def reset(self):
        self.count=0
        
#init a class
clicker1 = CounterClicker() #init to the default value of zero
clicker2 = CounterClicker(100) #setting the init count to 100
clicker3 = CounterClicker(count=100) #same but explicitly stating that we are setting the count argument to 100.

clicker4 = CounterClicker()

assert clicker4.read()==0, "clicker should start with zero as it wasn't given any argument values"
clicker4.click()
clicker4.click()
assert clicker4.read()==2, "after two clicks, clicker should have count 2"
clicker4.reset()
assert clicker4.read()==0, "after reset, clicker should have count 0"


#we can create subclass that inherit from parent class.

class NoResetClicker(CounterClicker):
    #this class has the same functions, except that we will override the old reset function
    def reset(self):
        pass


clicker5=NoResetClicker()
assert clicker5.read()==0
clicker5.click()
clicker5.click()
assert clicker5.read()==2
clicker5.reset()  #this time the NoResetClick reset function is call which just simple "pass".
assert clicker5.read()==2, "reset shouldn't do anything"
    
```

```python
#ITERABLES AND GENERATORS
"""are we using this so that we don't need to assign a variable and therefore store a list in memory for longer than the task it is needed for? so say we want to run something with every even number from 1-million, rather than storing that long list in a variable and therofre in memory we use a generator to on the fly produce this at the point of iterating over it in a for loop for example"""

#1. Create generators with functions using 'yield'
def generate_range(n):
    i=0
    while i<n:
        yield i #every call to yield, will cause the program to produce/yield another value of the generator according to the generator rule
        i +=1 #this is the rule used to get the next value to yield.
        
#then in a for loop, it will consume the yieldes valuves one at a time.

for i in generate_range(20):
    print(f"i: {i}")
    
    

#we can also generate infinite sequences with generators
def natural_numbers():
    n=1
    while True:
        yield n
        n += 1
        
#2. We can use comprehension to create the generators but instead of the square brackets we put then in parenthese ().

evens_below_20 = (x for x in generate_range(20) if x%2==0) #this generator comprehension does nothing until you iterate it

data = generate_range(40)
evens = (x for x in data if x%2==0)
evens_squares = (x*x for x in evens)
evens_squares_ending_in_six = (x for x in evens_squares if x%10==6)
#none of the above get created unitl the generator is referenced in the for or while

#as these are created on the fly len() will have no way of know the size of the object.

#often when iterating over a list we will want both the value and the index. Therefore we have enumerate() which turns values->(index, value) pair.
names = ['Edwin','Jonh','Karen']

for i, name in enumerate(names):
    print(f"name {i} is {name}")
```

```python
#Matrices
from typing import Tuple

#define a Matrix
Matrix = List[List[float]]

def shape(A: Matrix) -> Tuple[int,int]:
    """returns # of rows and columns of a Matrix"""
    rows = len(A)
    columns = len(A[0])
    return (rows, columns)


#define our A, B matrices
A: Matrix = [[1,2,3],
             [4,5,6]]
    
B: Matrix = [[1,2],
            [3,4],
            [5,6]] #3 rows, 2 columns
    
assert shape(A) == (2,3)

def get_row(A: Matrix, i: int) -> Vector:
    rows, columns = shape(A)
    assert i <= rows, "you're trying to access a column that doesn't exist"
    return A[i]

get_row(A,0)

def get_column(A: Matrix, i: int) -> Vector:
    rows, columns = shape(A)
    assert i <= columns, "you're trying to access a column that doesn't exist"
    return [A[j][i] for j in range(rows)]

get_column(A, 0)

from typing import Callable

#create a Matrix generator functions
def make_a_matrix(num_rows:int, num_col: int, entry_func:Callable[[int,int], float])->Matrix:
    """returns a matrix size num_rows x num_cols, whose entries (i,j)-th entry is entry_func(i,j)"""
    return [[entry_func(i,j)  #given a generator function, create a list or lists.
             for j in range(num_col)] 
            for i in range(num_rows)]

#demo create an identity matrix with 1s on the diagonal.
make_a_matrix(5,5, lambda i,j: 1 if i==j else 0) #here we will use a lambda function to pass through to create the matrix

#we could also define a matrix null (all zeros)
def my_func(i:int ,j:int)->float:
    return 1 if i==j else 0

make_a_matrix(6,6,my_func)

#or something more interesting
def my_func2(i:int, j:int) -> float:
    return i*math.sin(90*j)

make_a_matrix(4,4,my_func2)
```

```python
#STATISTICS
#used to better understand data. We can use it to distill and communicate relevant features of our data.
def mean(v: Vector) -> float:
    """compute the mean: sum of the values / number of elements"""
    return sum(v_i for v_i in v)/len(v)

mean(v)

#now we can create a function for the median
def median(v: Vector) -> float:
    """if the number of elements is odd, then take the middle value as the median, 
    otherwise, it is the average of the two middle elements"""
    
    sorted_vector = sorted(v) #we must first sort the list
    num_of_elements = len(v)
    mid_point = (num_of_elements)//2
    if num_of_elements%2==0: #even
        #print(f"even so use these values {sorted_vector[(num_of_elements-1)//2]} and {sorted_vector[(num_of_elements)//2]} ")
        return (sorted_vector[mid_point-1]+ sorted_vector[mid_point])/2
    else:
        return sorted_vector[mid_point]

assert median([1,1,2,2,3,3,4,5,5,6,100])== 3


#the quantile: remember that the median is just the q=0.5 (reprensenting the value under which 50% of the data lies).
def quantile(v: Vector, q: float) -> float:
    """compute the quantile
    1. sort the list
    2. get the q-quantile value"""
    
    return sorted(v)[int(q*len(v))]

quantile([1,1,2,2,3,3,4,5,5,6,100],0.5)

from collections import Counter
#the mode: the most freqent data point
def mode(v: Vector) -> Vector:
    counts = Counter(v)
    max_count = max(counts.values())
    return [v_i for v_i, count in counts.items() if count==max_count]
    
mode([1,1,1,2,3,4,5,5,5])
```

```python
#DISPERSION: the spread of the data

def range_of_data(v: Vector) -> float:
    #get the biggest and smallest elements and then get the difference
   # return sorted(v)[-1]-sorted(v)[0] #could just use the max and min func. probably faster than doing two sorts
    return max(v)-min(v)

range_of_data(v)

def de_mean(v: Vector) -> Vector:
    x_bar = mean(v) #subtract the mean so that the mean has zero value.
    return [x_i-x_bar for x_i in v]

#variance
def variance_of_data(v: Vector) -> float:
    """The sample variance: almost the average squared deviation/difference from the mean"""
    assert len(v) >=2, "variance requires at least two elements"
    #find the mean
    x_bar=mean(v)
    n = len(v)
    #find the sum of the squared differences between the vector and the value X-bar.
    value = sum((x_i-x_bar)**2 for x_i in v)
    return value/(n-1)

variance_of_data([1,2,3,4,5])

def standard_deviation_of_data(v: Vector) -> float:
    """the standard deviation is the sqrt of the variance"""
    return math.sqrt(variance_of_data(v))

standard_deviation_of_data([1,2,3,4,5])

def IQR(v: Vector)->float:
    return quantile(v,0.75) - quantile(v, 0.25)

IQR([1,2,3,4,5])
```

```python
#CORRELATION - The relationship between data.

#1. Covariance - the paired analogue of variance. Whereas variance measures how a single variable deviates from its mean,
#covariance measures how two variables vary in tandem from their means.

def covariance(v: Vector, w: Vector) -> float:
    assert len(v)==len(w), "v and w must have the same number of elements"
    
    return vector_dot_product(de_mean(v),de_mean(w))/(len(v)-1) #this is essentially sum of (i=0 to i=n) (x_i-x_bar)*(y_i-y_bar)/(n-1)
 
covariance(v,w)

def correlation(v: Vector, w: Vector) -> float:
    std_v=standard_deviation_of_data(v)
    std_w=standard_deviation_of_data(w)
    
    return covariance(v,w)/std_v/std_w

a=[1,2,3,4,5,6,12,1,2,3,4,6]
b=[2,-1,2,4,5,3,2,5,6,7,2,3]
correlation(a,b)
```
